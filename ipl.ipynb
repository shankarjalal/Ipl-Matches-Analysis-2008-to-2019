# **Ipl Matches Analysis 2008 to 2019**

```python
## Loading the required libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

set = sns.set_style('whitegrid')

import warnings 
warnings.filterwarnings('ignore')
```

```python
# Loading the ipl matches dataset.
ipl = pd.read_csv('c:/Users/DELL/Downloads/ipl_matches.csv', parse_dates = ['date'])
```

```python
## having a glance at the first five records of the dataset.
ipl.head()
```

```python
## Getting the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5] 
```

```python
## Getting the top 5 players with most man of the match awards.
ipl['player_of_match'].value_counts()[0:3]
```

```python
man_match = ipl['player_of_match'].value_counts()[0:5]
print(man_match)
plt.figure(figsize=(8,3))
sns.barplot(x=man_match.index, y=man_match.values, palette="viridis")
plt.title('Top 5 Players with Most Man of the Match Awards')
plt.ylabel('Number of Awards')
plt.xlabel('Player')
plt.show()
```

```python
# Making a bar plot for the top 5 players with man of the match awards.

plt.figure(figsize=(6,3))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),
         list(ipl['player_of_match'].value_counts()[0:5]),
         color= 'Blue',width = 0.4)
plt.title('Top 5: Most Man of the Match ')
plt.show()

# Max matchs ch gayle, ab d, and Rohit Sharma were man of the match.
```

```python
## Getting the frequency of result column.
ipl['result'].value_counts()
```

```python
plt.figure(figsize=(15, 5))
s = sns.countplot(x = 'toss_winner', data = ipl, palette= 'cool')
s.bar_label(s.containers[0])
plt.title('')
plt.xticks(rotation = 50)
plt.show()

# max time MI and KKR wining the toss.
```

```python
# Finding out the number of tross wins team

ipl['toss_winner'].value_counts().plot(kind = 'barh', color = 'red')
plt.title('Max time toss winner')
plt.show()
```

```python
# Extracting the records where a team won batting first.
batting_first = ipl[ipl['win_by_runs'] != 0]
```

```python
## making a histogram 

plt.Figure(figsize=(2,2))
sns.histplot(batting_first['win_by_runs'], bins = 30, kde= True,color = 'green')
plt.title('Won By Runs')
plt.xlabel('Runs')
plt.show()

# Mostly Teams are wining in less then 20 runs
```

```python
## finding out the number of wins each team after batting first.
batting_first['winner'].value_counts()[0:5]
```

```python
## Making a bar-plot for top 5 teams with most wins after batting first .

plt.figure(figsize=(7,3))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()), list(batting_first['winner'].value_counts()[0:5]), color=['blue', 'yellow', 'orange', 'purple', 'red'])
plt.title('Batting First With Win Matches')
plt.xticks(rotation = 60)
plt.show()

## And I get max matches won MI and CSK in IPL HIstory.
```

```python
ipl.head(1)
```

```python
count_match = ipl['winner'].value_counts(normalize=True).head(5)

plt.figure(figsize=(3,3))
plt.pie(count_match, labels = count_match.index, autopct = '%1.2f%%',  startangle = 90)
plt.title("Pie Plot With Winners")
plt.show()
```

```python
## Making  a Pia Chart

plt.Figure(figsize=(5,5))   # First perameter = numeric. Second perameter = categorical Values.
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.title('Batting First Wining Teams')
plt.show()

# MI has won most matches in Ipl by batting first.
```

```python
# extracting those records where a team has won after batting second.

batting_second = ipl[ipl['win_by_wickets'] != 0]
```

```python
# Making a histogram for frequencey of win number of wickets

plt.Figure(figsize=(4,4))
plt.hist(batting_second['win_by_wickets'], bins = 20, color='green')
plt.title('Betting Second for Win Matches')
plt.xlabel('Wickets')
plt.show()

# While batting second I analyzed that most of the teams won the match by 6 & 8 wickets.
```

```python
## finding out the frequency of numbers of wins each time after batting second.
batting_second['winner'].value_counts()
```

```python
# Making a pie chart for distribution of most after batting second.
plt.Figure(figsize=(4,4))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()), autopct= '%1.01f%%')
plt.show()

# KKR has won most matches in Ipl by batting Second.
```

```python
# Looking at the number of matches pleyed each season

yearly_count = ipl['season'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,3))
plt.plot(yearly_count.index, yearly_count.values, marker = 'o', linestyle = '-', color = 'darkblue')
plt.title('Number of Matches Played per Season.')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.ylim(20,80)
plt.grid(axis = 'y')
plt.show()
```

```python
## Looking at the number of matches played in each city

ipl['city'].value_counts().head(10).plot(kind = 'bar')
plt.title('Top 10 Maximum Matches Played City')
plt.show()
```
